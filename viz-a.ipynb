```python
import seaborn as sns
import pandas as pd
```

```python
sns.set()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.info()
```

#### The tip increases when the bill is higher

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

#### Males pay more for their meals than females

```python
sns.barplot(x = "time", y= "total_bill", hue = "sex", data = tips, ci=None);
```

##### Sequentially, it shows that men tip more than females on both dinner and lunch

```python
sns.barplot(x = "time", y= "tip", hue = "sex", data = tips, ci=None);
```

#### Conclusion: there is a correlation between the bill and the amount of the tip. The higher the bill, the higher the tip gets. Males tend to tip more because they spend more money on their meals.
